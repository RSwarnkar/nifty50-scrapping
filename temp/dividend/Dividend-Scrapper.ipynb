```python
import bs4 as bs
from bs4 import BeautifulSoup

data = []
soup = BeautifulSoup(rs.text, 'html.parser')
table = soup.find('table', attrs={'class':'mctable1'})
table_body = table.find('tbody')

rows = table_body.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele]) # Get rid of empty values
```

```python
df_links = pd.read_csv('../../configs/moneycontrol-dividend-links.csv')
df_links.head()
```

```python
import time


path = './div/'
for index, row in df_links.iterrows():
    #print(row['STOCK'], row['URL'])
    stock_name = row['STOCK']
    url = row['URL']
    all_tables = pd.read_html(url)
    #print(f'Total tables: {len(all_tables)}')
    df_temp = all_tables[0]
    df_temp['Stockname'] = stock_name
    
    df_temp['Announcement Date'] = pd.to_datetime(df_temp['Announcement Date'])
    df_temp['Ex-Date']           = pd.to_datetime(df_temp['Ex-Date'])
    
    df_temp['Announce Date']     = df_temp['Announcement Date'].dt.strftime('%m/%d/%Y')
    df_temp['Ex Date']           = df_temp['Ex-Date'].dt.strftime('%m/%d/%Y')
    
    df_temp.to_csv(str(path+stock_name+'.csv'), index=False)
    time.sleep(1)
    print("Done: "+stock_name)
    
    
    
    
```
