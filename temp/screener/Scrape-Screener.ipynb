```python
stocklistfile = "../../configs/common-stocklist.csv"
stocklist = pd.read_csv(stocklistfile)
stocklist.head()
```

```python
import time


path = "./holdings/"
for index, row in stocklist.iterrows():
    stock_name = row['STOCK']
    
    url = "https://www.screener.in/company/"+stock_name+"/"
    
    all_tables = pd.read_html(url)
    #print(f'Total tables: {len(all_tables)}')
    df_temp = all_tables[-1] # 9th Table is holdings which is also last table 
    df_temp['Stockname'] = stock_name

    df_temp.to_csv(str(path+stock_name+'.csv'), index=False)
    time.sleep(1)
    print("Done: "+stock_name)
    
```

```python
import time
import requests as rq
import pandas as pd
import numpy as np 


stocklistfile = "../../configs/common-stocklist.csv"
stocklist = pd.read_csv(stocklistfile)

path = "./cfoa/"
for index, row in stocklist.iterrows():
    stock_name = row['STOCK']
    
    url = "https://www.screener.in/company/"+stock_name+"/"
    
    all_tables = pd.read_html(url)
    #print(f'Total tables: {len(all_tables)}')
    df_temp = all_tables[7] # 7th Table is Cashflow from operating activity  
    df_temp['Stockname'] = stock_name

    df_temp.to_csv(str(path+stock_name+'.csv'), index=False)
    time.sleep(1)
    print("Done: "+stock_name)
```
